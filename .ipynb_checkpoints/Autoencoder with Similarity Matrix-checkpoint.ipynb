```python
#datapath = "Sample Data"
names_datapath = "E:\\Development Project\\Data\\GNPS" #Absolute path for data.
datapath = "E:\\Development Project\\Data\\GNPS Python Master\\Final Data.txt"
fingerprints_path = "E:\\Development Project\\Data\\GNPS Python Master\\Final Fingerprints.txt"
fingerprints_names_path = "E:\\Development Project\\Data\\GNPS Python Master\\Fingerprint Legend.txt"
similarities_path = "E:\\Development Project\\Data\\Pairwise Spectra Similarities\\matej_sim.csv"
num_samples = 5770
num_val_samples = 1000
numFeatures = 2995
batch_size = 500
default_dpi = plt.rcParamsDefault['figure.dpi']
plt.rcParams['figure.dpi'] = default_dpi*1.1
```

```python
def load_master_file():
    MAX_MASS = 1000
    BIN_SIZE = 1
    
    mol_all = np.loadtxt(datapath, dtype="U25") # Get master file in as numpy array
    
    mol_ids = np.unique(mol_all[:, 0])  # Trim duplicate filename rows.
    
    # Construct empty Pandas dataframe of correct size.
    intensities = pd.DataFrame(0.0, index = mol_ids, columns=range(MAX_MASS//BIN_SIZE), dtype=float)
    print(intensities.index)
    
    # Populate the dataframe.
    for row in mol_all:
        intensities.at[row[0], float(row[1])-1] = float(row[2])
    
    # Convert populated dataframe into a numpy array for use by autoencoder
    np_matrix = intensities.values
    return np_matrix

def load_similarity_matrix():
    headers = np.zeros(num_samples, dtype="U25")
    # Get file names to use as headers.
    for i, file in enumerate(os.listdir(names_datapath)): 
        headers[i] = file
    # Create empy dataframe to hold matrix, read in similarity values.
    similarities = pd.DataFrame(0, index=headers, columns=headers, dtype=float)
    raw_similarities = pd.read_csv(similarities_path, header=None, usecols=[0,1,3])
    # Insert similarity value for molecule pair in correct cell.
    for row in raw_similarities.itertuples(index=True, name='Pandas'):
        mol_1 = row[1]
        mol_2 = row[2]
        value = row[3]
        similarities.at[mol_1, mol_2] = value
        
    # Turn into a 2D numpy array
    np_form = similarities.values
    
    # Find rows where all values are 0, change to 1/num_samples to avoid /0
    sums = np.sum(np_form, axis=1)
    np_form[np.where(sums==0)] = 1/num_samples
    
    # Turn into a 3D numpy array, representing batches
    
    num_batches = int(num_samples/batch_size)
    
    np_matrix = np.zeros((num_batches, batch_size, batch_size))
    
    for i, start in enumerate(range(0, num_samples-batch_size+1, batch_size)):
        np_matrix[i] = np_form[start:start+batch_size, start:start+batch_size]
        np_matrix[i] = 0.5 * (np_matrix[i] + np_matrix[i].T) # Make symmetric?
        np_matrix[i] = np_matrix[i]/np_matrix[i].sum(axis=1)[:, None] # Normalize by row.
    
    # np_matrix = 0.5 * (np_matrix + np_matrix.T) # Make symmetric?

    # print(sums)
    # np_matrix = np_matrix/np_matrix.sum(axis=1)[:, None] # Normalize by row.
    # sums = np.sum(np_matrix, axis=1)
    # print(np.where(sums<0.99)) # Verify all rows sum up to 1.
    
    return np_matrix

def load_similarity_matrix_by_batch():
    headers = np.zeros(num_samples, dtype="U25")
    # Get file names to use as headers.
    for i, file in enumerate(os.listdir(names_datapath)):
        headers[i] = file
    # Create empy dataframe to hold matrix, read in similarity values.
    similarities = pd.DataFrame(0, index=headers, columns=headers, dtype=float)
    raw_similarities = pd.read_csv(similarities_path, header=None, usecols=[0, 1, 3])
    # Insert similarity value for molecule pair in correct cell.
    for row in raw_similarities.itertuples(index=True, name='Pandas'):
        mol_1 = row[1]
        mol_2 = row[2]
        value = row[3]
        similarities.at[mol_1, mol_2] = value

    # Turn into a 2D numpy array
    np_form = similarities.values

    # Find rows where all values are 0, change to 1/num_samples to avoid /0
    # Should this be done on a whole-matrix scale or batch-by-batch
    # sums = np.sum(np_form, axis=1)
    # print(np.where(sums == 0))
    # np_form[np.where(sums == 0)] = 1 / num_samples

    # Turn into a 3D numpy array, representing batches
    num_batches = int(num_samples / batch_size)

    np_matrix = np.zeros((num_batches, batch_size, batch_size))

    for i, start in enumerate(range(0, num_samples - batch_size + 1, batch_size)):
        np_matrix[i] = np_form[start:start + batch_size, start:start + batch_size]
        sums = np.sum(np_matrix[i], axis=1)
        np_matrix[i][np.where(sums == 0)] = 1 / batch_size
        np_matrix[i] = 0.5 * (np_matrix[i] + np_matrix[i].T)  # Make symmetric?
        np_matrix[i] = np_matrix[i] / np_matrix[i].sum(axis=1)[:, None]  # Normalize by row.
    
    return np_matrix
    
```

```python
def tsne(P, activations):
    d = 16 # TODO: should set this automatically
    n = 500 #batch_size # TODO: should set this automatically. I believe 32 is the default used by Keras
    v = d - 1.
    eps = K.variable(10e-15) # needs to be at least 10e-8 to get anything after Q /= K.sum(Q)
    sum_act = K.sum(K.square(activations), axis=1)
    Q = K.reshape(sum_act, [-1, 1]) + -2 * K.dot(activations, K.transpose(activations))
    Q = (sum_act + Q) / v
    Q = K.pow(1 + Q, -(v + 1) / 2)
    Q *= K.variable(1 - np.eye(n))
    Q /= K.sum(Q)
    Q = K.maximum(Q, eps)
    C = K.log((P + eps) / (Q + eps))
    C = K.sum(P * C)
    return C
```

```python
def plot_loss(fitted_model, epochs):
    loss = fitted_model.history['loss']
    val_loss = fitted_model.history['val_loss']
    epochs_label = epochs
    epochs = range(epochs)

    plt.figure()
    plt.plot(epochs, loss, 'bo', label='Training loss')
    plt.plot(epochs, val_loss,'r', label='Validation loss')
    plt.title('Training and Validation Loss for ' + str(epochs_label) + ' epochs')
    plt.legend()
    plt.show()
```

```python
from keras.optimizers import SGD
def basic_autoencoder(x_train, epochs=100, encoded_dim=10):
    inputLayer = Input(shape=(x_train.shape[1],))  # fixed
    output_dims = x_train.shape[1]

    l = inputLayer
    l = Dense(256, activation='relu')(l)
    l = Dense(128, activation='relu')(l)
    l = Dense(16, activation='relu')(l)
    # l = Dense(encoded_dim, activation='relu')(l)

    latent_space = l

    #l2 = Dense(16, activation='relu')(l)
    l2 = Dense(128, activation='relu')(l)
    l2 = Dense(256, activation='relu')(l2)
    l2 = Dense(output_dims, activation='relu')(l2)

    out_layer = l2

    auto_model = Model(input=inputLayer, output=out_layer)  # Create model.
    
    # encoder = Model(inputs=inputLayer, outputs=latent_space)
    
    # encoded_input = Input(shape=(encoded_dim,))
    # decoder_layer = auto_model.layers[-4]
    # decoder = Model(encoded_input, decoder_layer(encoded_input))
    
    sgd = SGD(lr=0.1)
    auto_model.compile(loss='mean_squared_error', optimizer=sgd)

    # Fixed this line: added x_train again: you need to give it the targets! Also removed the batch stuff
    #autoencoder_train = auto_model.fit(x_train, x_train, shuffle=False, epochs=epochs, validation_split = 0.1)
    autoencoder_train = auto_model.fit(x_train, x_train, shuffle=False, validation_split = 0.1, epochs=epochs)
    
    plot_loss(autoencoder_train, epochs)
    
    return auto_model  # Return the model, now trained.
```

```python
def similarity_autoencoder(x_train, y_train, epochs=100, encoded_dim=10):
    input_layer = Input(shape=(x_train.shape[1],))  # fixed
    output_dims = x_train.shape[1]

    l = input_layer
    l = Dense(256, activation='relu')(l)
    l = Dense(128, activation='relu')(l)
    
    l = Dense(16, activation='relu')(l)
    # l = Dense(encoded_dim, activation='relu')(l)

    latent_space = l

    #l2 = Dense(16, activation='relu')(l)
    l2 = Dense(128, activation='relu')(l)
    l2 = Dense(256, activation='relu')(l2)
    l2 = Dense(output_dims, activation='relu')(l2)
    
    out_layer = l2
    
    auto_model = Model(input=input_layer, output=[out_layer, latent_space])
    
    sgd = SGD(lr=0.1)
    auto_model.compile(loss=['mean_squared_error',tsne], optimizer=sgd)
    
    autoencoder_train = auto_model.fit(x_train,[x_train, y_train], shuffle=False, validation_split = 0.0909, epochs=epochs, batch_size=batch_size)
    
    plot_loss(autoencoder_train, epochs)
    
    return auto_model
    
```
